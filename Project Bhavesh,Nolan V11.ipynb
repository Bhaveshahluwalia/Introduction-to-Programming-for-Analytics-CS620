```python
# this function takes state name as an argument and plot the line graph with frequency and date 
def state_posting(state):
    '''prints a line graph for job postings over time by state, where the user needs to input a state by its 2 letter abbreviation'''
    state = df.loc[df['state']== state] # here we adding the company name which function take it as a argument
    state =state['post_date'].str[:-3] #adding the dates in the company
    state.value_counts() #counting the name of the company from dataframe
    state.to_frame() # #making new frame for the particular company
    state =  state.value_counts().reset_index() #creating new dataframe that adds counts
    state.columns = ['post_date', 'count'] #calling our own columns we want
    state =  state.sort_values(by=['post_date']) #sorting data in date order
    state.plot(figsize = (20,10),x="post_date",y="count", color='green', linestyle='dashed', linewidth = 3, 
         marker='o', markerfacecolor='blue', markersize=12)
 #ploting the graph for the particular company
    plt.xlabel("Date",size=20)   # "Date" label on the x axis
    plt.ylabel("Frequency",size=20) #"Frequency" label on y axis
    plt.title("state with post date", size=20) #Plot Graph title 
    return 
```

```python
#Here we create a function using our newly created list for job_description
def common_words(i,lst): #I equals the amount of common words we want to see. lst = the list we want to find common words in
    from collections import Counter #importing counters to get our frequency counts
    count = Counter(word for word in lst) #now counting for each word
    return count.most_common(i) #return our response back to back to function
```
